```python
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir('/Users/ephraimlove/Courses/data_wrangling')

df = pd.read_csv('data/assignment_5/raw/mpg.csv')

summary_all = df.describe(include='all')
print(summary_all)
```

```python
import seaborn as sns

tips_df = sns.load_dataset('tips')
print(tips_df.head())
```

```python
tips_df['tip_pct'] = tips_df['tip'] / tips_df['total_bill']

# Create the box and whisker plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips_df, x='sex', y='tip_pct')

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Tip Percentage')
plt.title('Box and Whisker Plot of Tip Percentage by Sex')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create the box and whisker plot without outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips_df, x='sex', y='tip_pct', showfliers=False)

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Tip Percentage')
plt.title('Box and Whisker Plot of Tip Percentage by Sex (No Outliers)')

# Show the plot
plt.tight_layout()
plt.show()
```
